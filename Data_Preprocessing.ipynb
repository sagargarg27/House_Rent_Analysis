```python
df['companyName']=df.companyName.str.replace('A to Z Property','A_to_Z_Property').str.replace('A 2 Z Property','A_to_Z_Property').str.replace('A to z realty','A_to_Z_Property')
```

```python
df['companyName'].value_counts()
```

```python
comp_price = df[['companyName','price']]
comp_price
```

```python
cp = comp_price.groupby('companyName')['price'].mean()
cp.sort_values(ascending=False, inplace=True)
cp = cp.head(100)
cp
```

```python
ccp = pd.DataFrame(cp).reset_index()
companyNames = ccp['companyName'].to_list()
companyNames
```

```python
ACN=df['companyName'].values.tolist()
```

```python
for i in range(0,len(ACN)-1):
    if ACN[i] not in companyNames:
        ACN[i]='Other'
```

```python
new_company_names= pd.DataFrame(ACN)
df['companyName']=new_company_names
df.head()
```
